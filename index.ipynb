## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A SQL database organizes its data in table form. It would be used when you have different types of data that relates to eachother but isn't necessarily in the same category.

A NoSQL database organizes data in tree form. A document can contain documents and is not required to have a set schema.

Neither is better than the other, each would be used in a different situation.

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________



If a server is down, fault tolerance allows the running servers to access backups of the disabled server to complete the query.




## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MapReduce allows servers to run one query at the same time. If the data required for a query is distributed accross multiple servers at different locations, MapReduce pulls that data simultaneously which speeds up the query.
